# Task3: Let's find out how many NaN (Not a Number) are there in the dataset

### 1. We will be defining an empty list.
### 2. We will be running a for loop over each of the columns in the dataset.
### 3. For each of the column we will be using the pandas method .isna() to find out if a particular cell is empty or not.
### 4. For each column, the loop will scan through each cell and find out the total number of times it has encountered an empty cell and sum it.
### 5. Finally, it will give us in the form of a list empty cells for all the columns in the dataset.

```python
#Defining the empty list
lst_nan = []

for i in range(len(df.columns)):
    lst_nan.append(df.iloc[:,i].isna().sum())
```

```python
#Printing the list
lst_nan
```

```python
#Converting that into a dataframe
df_nan = pd.DataFrame(lst_nan,index = df.columns)
```

```python
df_nan
```

# Conclusion

## 1. There are no empty values in the dataset. This was noted from the df.info() line however, this approach is typically useful in the case where one needs to identify the null values for each column.

# Task4: Let's find out how many unique values are there in the dataset

### 1. We will be defining an empty list.
### 2. We will be running a for loop over each of the columns in the dataset.
### 3. For each of the column we will be using the pandas method .nunique() to find out how many unique values are there in that column.
### 4. For each column, the loop will scan through each cell and find out the total number of times it has encountered an unique value.
### 5. Finally, it will give us the number of unique values in the form of a list for all the columns in the dataset.

```python
#Defining the empty list
lst_unique = []

for i in range(len(df.columns)):
    lst_unique.append(df.iloc[:,i].nunique())
```

```python
#Printing the list
lst_unique
```

```python
#Converting that into a dataframe
df_unique = pd.DataFrame(lst_unique,index = df.columns)
```

```python
df_unique
```

# Conclusion

## Unique values help the analyst to determine whether a particular variable (column) should be treated as a continous variable or a categorical variable. 

## Lower unique values mean that using that particular variable the data can be divided into categories and vice versa. Typically true for variables such as gender, region etc.

## 1. Highest number of unique values are in the charges column. This means this can be treated as a continuous variable.
## 2. Lowest number of unique values are in the sex and the smoker column followed by region. This means these can be treated as a categorical variable.
## 3. The number of children can also be treated as a categorical variable.
## 4. The bmi having higher number of unique values can be treated as a continuous variable.
